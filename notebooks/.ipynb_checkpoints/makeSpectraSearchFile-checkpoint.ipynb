```python
import numpy as np
```

```python
iD = np.genfromtxt('ID.dat', dtype = str)
alt_iD = np.genfromtxt('NAME.dat', dtype = str, delimiter='\n')
```

```python
iD
```

```python
with open('IDspectra.txt', 'w') as f:
    for i in range(len(iD)):
        if iD[i] == '-1':
            name = alt_iD[i]
            f.write('{}\n'.format(name))
        else:
            name= iD[i]
            f.write('2MASS {}\n'.format(name))    
```
